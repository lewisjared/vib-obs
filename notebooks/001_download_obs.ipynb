# Downloading the latest obs

BOM makes it's latest obs available on it's FTP server as an XML file for each state. There are hundreds of other files which contain other output, but for now we are interested in the wind observations over VIC.

Specifically the file of interest for Victoria is:

```
ftp://ftp2.bom.gov.au/anon/gen/fwo/IDV60920.xml
```

Let's download that (might be easier to do using wget...)

````python
from ftplib import FTP
````

````python
OUT_FILE = "../data/IDV60920.xml"
````

````python
FTP_HOST = "ftp2.bom.gov.au"
FTP_DIR = "anon/gen/fwo"
FTP_FILE = "IDV60920.xml"
````

````python
ftp = FTP(FTP_HOST) 
ftp.login()
````

````python
ftp.cwd(FTP_DIR)
````

````python
ftp.retrlines('LIST')
````

````python
with open(OUT_FILE, 'wb') as fp:
    ftp.retrbinary('RETR ' + FTP_FILE, fp.write)
````
